```python
from pandas import *
from matplotlib.pyplot import *
from seaborn import *
```

```python
p=read_csv("preg_demographs.csv")
p.head()
```

```python
p=p.drop(["Aadhaar_Number","Gender","Phone_Number","Email","Address", "Tests_Conducted","Pincode","Hospital_Name","Checkup_Date","Checkup_Time","Blood_Pressure", "User_Registration_Date","User_Registration_Time","Medical_History","Medications","Nutrition_Plan","Exercise_Routine","Vaccination_Records","Partner_Information","Ultrasound_Images","Insurance_Information","Delivery_Date","Successful_Checkup_Date","Successful_Checkup_Time","Disease","Observation","Suspects","Reminder_Date"], axis=1)
p
```

```python
p=p.sort_values(["Age"])
p
```

```python
p.to_csv("preg_demograph_2.csv", index=False)
p=read_csv("preg_demograph_2.csv")
```

```python
p.info
```

```python
kdeplot(data=p, x='Fetal_Heart_Rate', bw_adjust=0.5, fill=True, color='green', linestyle='--', alpha=0.5)
title("Kdeplot")
```

```python
figure(figsize=(50,40))
p.plot.hexbin(x='YOB', y='No_of_Checkups', gridsize=10, cmap='coolwarm')
title("Hexbin plot")
```

```python
lineplot(data=p, x='Weight(kg)', y='Gestational_Age')
title("Line plot")
```

```python
c_palette=["#E78A9A","#DB2AAF"]
countplot(x="Baby_Gender", hue="Delivery_Type", palette=c_palette, data=p)
title("Count plot")
```

```python
c_palette=["#D20103","#FFECA1"]
countplot(x="Mother_Blood_Type", hue="Anomaly", palette=c_palette, data=p)
title("Count plot")
```

```python
figure(figsize=(15,6))
barplot(x="Gestational_Age", y="No_of_Missed_Checkups", hue="Maternal_Mental_Health", data=p)
title("Bar plot")
```

```python
figure(figsize=(8,5))
violinplot(x="Mother_Blood_Type", y="Fetal_Heart_Rate", data=p, color="0.8")
stripplot(x="Mother_Blood_Type", y="Fetal_Heart_Rate", data=p, jitter=True)
title("Stripplot with Violin plot")
```

```python
displot(p['Age'], kde=True, bins=15)
title("Displot")
```
